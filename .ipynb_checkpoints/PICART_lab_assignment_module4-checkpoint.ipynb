### PROBLEM 1
**The Lotka-Volterra equations**

_Refer to Exercise 8.2 in the Newman text._ The Lotka-Volterra equations are a mathematical model of predator-prey interactions between biological species. Let two variables $x$ and $y$ be proportional to the size of the populations of two species, traditionally called "rabbits" (the prey) and "foxes" (the predators). You could think of $x$ and $y$ as being the population in thousands, say, so that $x = 2$ means there are 2000 rabbits.

In the Lotka-Volterra model the rabbits reproduce at a rate proportional to their population, but are eaten by the foxes at a rate proportional to both their own population and the population of foxes:

$$\frac{{\rm d}x}{{\rm d}t} = \alpha x - \beta xy,$$

where $\alpha$ and $\beta$ are constants. At the same time the foxes reproduce at a rate proportional to the rate at which they eat rabbits-- because they need food to grow and reproduce-- but also die of old age at a rate proportional to their own population:

$$\frac{{\rm d}y}{{\rm d}t}  = \gamma xy - \delta y,$$

where $\gamma$ and $\delta$ are also constants. *(25 pts.)*

1. Solve these equations using the fourth-order Runge-Kutta method for the case $\alpha = 1, \beta = \gamma = 0.5$, and $\delta = 2$, starting from the initial condition $x = y = 2$.

2. Make a graph showing both $x$ and $y$ as a function of time on the same axes from $t =0$ to $t = 30$. *(25 pts.)*

```python
import numpy as np
import matplotlib.pyplot as plt

alpha, beta, gamma, delta = 1, 0.5, 0.5, 2

def f(r):           #functions are not dependent on time, input is an array
  x = r[0]          #first element is x value
  y = r[1]          #second element is y value
  fx = alpha*x - beta*x*y         #dx/dt
  fy = gamma*x*y - delta*y        #dy/dt
  return np.array([fx,fy], float)        #return an array with [dx/dt,dy/dt]

a = 0         #time endpoints
b = 30
N = 300       #number of steps
h = (b-a)/N   #size of step

tpoints = np.arange(a,b,h)          #time array
xpoints = []                        #initialize x and y lists
ypoints = []
r = np.array([2,2], float)          #initial values for x=y=2

for t in tpoints:
  xpoints.append(r[0])              #store current value for x and y
  ypoints.append(r[1])
  #4th order RK method
  k1 = h*f(r)           #f(r) is the slope at (t,r), *h gives the ~ step in r after h
  k2 = h*f(r+0.5*k1)    #f term is slope at the midpoint between t, t+h, with r value estimated by k1
  k3 = h*f(r+0.5*k2)    #f term is again slope at midpoint, but with r value estimated by k2
  k4 = h*f(r+k3)        #f(r+k3) is now the slope at t+h, with r value estimated by k3
  r += (k1+2*k2+2*k3+k4)/6      #similar to trapezoidal method, each midpoint value gets twice the weight, and /6 to normalize


plt.plot(tpoints, xpoints, label = "Rabbits")
plt.plot(tpoints, ypoints, label = "Foxes")
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend()
plt.title("Plot")
plt.show()
```

For the 4th order RK method, the two midpoints will tend to overshoot the value. Assuming the function's derivative does not drastically change within the interval t->t+h (this can be done by taking smaller time steps), the two midpoint values will sandwich the solution function. This results in a faster convergence than choosing only k2, which is in the midpoint method.

Result: a periodic function that
*   after t=0: $x' > 0, y' < 0$ \\
      starts with rabbits reproducing rapidly and foxes dying due to low prey numbers and old age.
*   $x' > 0, y'< 0$ \\
      As the rabbits population increases, the foxes graph concaves upward signifying that foxes are starting to be able to eat.
*   $x' < 0, y'> 0$ \\
      After the rabbit population peaks, the foxes number starts to concave downward. Foxes are still able to eat, but are starting to slow down.
*   $x' < 0, y'< 0$ \\
      After the foxes number peaks, the rabbits are decreasing drastically as they are preyed upon. The decreasing rabbits eventually leads to decreasing foxes until the system resets with the initial values.

Increasing the parameters can change the system by the following: \\
1. $\alpha$ decreases the period of the function and increases amplitude of the
peaks, though it favors foxes ($\alpha$ can be interpreted as how fast the rabbits reproduce. Its direct effect on the rabbit population effectively has a bigger change in xy, resulting in a "sped up" version of the system. It has a similar effect with increasing $\gamma$.) \\
2. $\beta$ allows a higher rabbit peak, lower fox peak (equating x'=0 gives $\beta xy = \alpha x$, meaning the peak in x is higher.) \\
3. $\gamma$ allows higher fox peak, lower rabbit peak (equating y'=0 gives $\gamma xy = \delta y$, meaning the peak in y is higher.) \\
4. $\delta$ increases the amplitude and sharpness of both peaks, though it favors rabbits ($\delta$ can be interpreted as how many foxes die of old age. It has a similar effect with increasing $\beta$.)\\

### PROBLEM 2

**Nonlinear pendulum**

_Refer to Exercise 8.4 in the Newman text._ Building on the results from Example 8.6, calculate the motion of a nonlinear pendulum as follows.

1. Solve the two first-order equations, Eqs. (8.45) and (8.46), using the fourth-order Runge-Kutta method for a pendulum with a 10 cm arm. Use your program to calculate the angle $\theta$ of displacement for several periods of the pendulum when it is released from a standstill at $\theta = 179^\circ$ from the vertical. *(25 pts.)*

2. Make a graph of $\theta$ as a function of time. *(25 pts.)*


```python
import numpy as np
import matplotlib.pyplot as plt

g = 9.8  #m/s^2
l = 0.1  #10cm arm length in m

def f(r):           #functions are not dependent on time, input is an array
  theta = r[0]          #first element is theta value
  omega = r[1]          #second element is omega value
  ft = omega         #d(theta)/dt
  fo = -g/l*np.sin(theta)        #d(omega)/dt
  return np.array([ft,fo], float)        #return an array with [theta',omega']

a = 0         #time endpoints
b = 5        #choose so that a few periods are visible
N = 500      #number of steps
h = (b-a)/N   #size of step

tpoints = np.arange(a,b,h)          #time array
thetapoints = np.array([], float)               #initialize theta list, using numpy array so the radians->degrees conversion is easier
r = np.array([179/180*np.pi,0], float)          #initial values for x=y=2

for t in tpoints:
  thetapoints = np.append(thetapoints, r[0])    #use numpy append to keep the float data type
  #4th order RK method
  k1 = h*f(r)           #f(r) is the slope at (t,r), *h gives the ~ step in r after h
  k2 = h*f(r+0.5*k1)    #f term is slope at the midpoint between t, t+h, with r value estimated by k1
  k3 = h*f(r+0.5*k2)    #f term is again slope at midpoint, but with r value estimated by k2
  k4 = h*f(r+k3)        #f(r+k3) is now the slope at t+h, with r value estimated by k3
  r += (k1+2*k2+2*k3+k4)/6      #similar to trapezoidal method, each midpoint value gets twice the weight, and /6 to normalize

plt.plot(tpoints, thetapoints*(180/np.pi), label = "θ")        #numpy array * float is not possible for a list
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Plot of theta with initial value 179 degrees, l=10cm")
plt.show()
```

For small angles, the graph is approximated by sines and cosines. This is visible near the peaks, except since 180 is an unstable equilibrium, it slowly goes away from the value instead of oscillating around it, explaining why it is almost flat at the peaks. (In these values near $\theta\approx180^\circ$, $\sin{\theta} = -\sin(180^\circ-\theta) = $, where the term inside the sine is actually small, thus can be approximated with small angles) but for large angles (around $\theta\approx90^\circ$), the higher order terms become relevant. \\

For the non-small angle approximation, it is easier to think of the sine term for what it is, the concavity of the graph. Initially, it gets more and more concave down, then starts decreasing around $\theta = 90^\circ$, then the inflection point is at $\theta = 0$, the concavity then flips as $\theta$ approaches $-179^\circ$. \\

As for the code itself, the accuracy can be set by the step size, since large step sizes will overshoot the values, leading to very large negative values. The finite step size combined with a large enough value for b (assuming the graph converges), shows the wave to dampen and eventually zero out. This is not ideal since it implies that energy is dissipated from the system, of which there are none in the equations.
