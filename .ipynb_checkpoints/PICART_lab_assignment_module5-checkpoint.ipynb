### PROBLEM 1
**The Schrodinger equation and the Crank-Nicolson method**

_Refer to Exercise 9.8 in the Newman text._ In this problem, you will use the Crank-Nicolson method to solve the full time-dependent Schrodinger equation and hence develop a picture of how a wavefunction evolves over time.

Consider an electron (mass $M = 9.109 \times 10^{-31}$ kg) in a box of length $L = 10^{-8}$ m. Suppose that at time $t = 0$ the wavefunction of the electron has the form

$$ \psi(x,0) = \exp\left[-\frac{(x-x_0)^2}{2\sigma^2}\right]e^{i\kappa x},$$
where $x_0 = \frac{L}{2}$, $\sigma = 1 \times 10^{-10}$ m, $\kappa = 5 \times 10^{10} {\rm m}^{-1}$,  and $\psi = 0$ on the walls at $x = 0$ and $x = L$.

1. Perform a single step of the Crank-Nicolson method for this electron, calculating the vector $\psi(t)$ of values of the wavefunction, given the initial wavefunction above and using $N = 1000$ spatial slices with $a = L/N$. Your program will have to perform the following steps. First, given the vector $\psi(0)$ at $t = 0$, you will have to multiply by the matrix $\bf{B}$ to get a vector $\bf{v} = \bf{B}\psi$. Because of the tridiagonal form of $\bf{B}$, this is fairly simple. The $i$th component of $\bf{v}$ is given by
$$ v_i = b_1\psi_i + b_2(\psi_{i+1} + \psi_{i-1}).$$

   You will also have to choose a value for the time-step $h$. A reasonable choice is $h = 10^{-18}$ s. *(30 pts.)*

2. Second you will have to solve the linear system ${\bf Ax}= {\bf v}$ for $\bf{x}$, which gives you the new value of $\psi$. You could do this using a standard linear equation solver like the function $\tt solve$ in numpy's $\tt linalg$. *(20 pts.)*

3. Once you have the code in place to perform a single step of the calculation, extend your program to perform repeated steps and hence solve for $\psi$ at a sequence of times a separation $h$ apart. Note that the matrix $\bf A$ is independent of time, so it doesn't change from one step to another. You can set up the matrix just once and then keep on reusing it for every step. *(30 pts.)*

4. Make an animation of the solution by displaying the real part of the wavefunction at each time-step. You can use the function rate from the package visual to ensure a smooth frame-rate for your animation-- see Section 3.5 on page 117 of the Newman text.

   Run your animation for a while and describe what you see. Write a few sentences explaining in physics terms what is going on in the system. *(20 pts.)*

```python
import numpy as np
import matplotlib.pyplot as plt         #used to make subplots
import matplotlib.animation             #used to draw the animation
from IPython.display import HTML        #used to embed animation

M = 9.109e-31      #mass of e-
L = 1e-8           #length of box
x_0 = L/2
sigma = 1e-10
kappa = 5e10
hbar = 1.0546e-34  #ħ reduced planck's constant

x0 = 0
xN = L
N = 1000       #spatial slices
a = L/N        #spatial step
h = 2e-18      #time step
frames = 5000   #total frames * time step = total duration (1e-14)
t = np.arange(0,frames*h,h)      #time array

def psi(x):                  #Ψ(x,0)
  return np.exp(-(x-x_0)**2/2/sigma/sigma)*np.exp(1j*kappa*x)

psi_x = np.zeros(N+1, dtype=complex)      #initialize Ψ_x0
for N in range(N+1):             #evaluate Ψ(x) for each x in 0 to L
  psi_x[N] = psi(N*a)
psi_xt = [np.real(psi_x)]        #take only real part

#from the textbook, elements of A and B from applying Crank-Nicolson method to SE specifically
a1 = 1 + h*1j*hbar/2/M/a/a
a2 = -h*1j*hbar/4/M/a/a
b1 = 1 - h*1j*hbar/2/M/a/a
b2 = h*1j*hbar/4/M/a/a

A = np.zeros((N+1,N+1), dtype=complex)        #initialize complex matrices
B = np.zeros((N+1,N+1), dtype=complex)

for i in range(N+1):
  A[i][i] = a1         #diagonal
  B[i][i] = b1

for i in range(N):
  A[i][i+1] = a2       #tridiagonal
  A[i+1][i] = a2
  B[i][i+1] = b2
  B[i+1][i] = b2

for i in range(frames):
  v = B @ psi_x                     #v = BΨ(x,t)
  psi_x = np.linalg.solve(A,v)      #AΨ(x,t+h) = v
  psi_xt.append(np.real(psi_x))     #take only real part again

#Initialize figs and axes for the animation
fig, ax = plt.subplots()
ax.axis([0,L,-1,1])
ax.set_xlabel('x')
ax.set_ylabel('Ψ(x)')
l, = ax.plot([],[])         #trailing comma unpacks the tuple from ax.plot, returning the plot itself instead of the ax.plot tuple

def animate(i):    #define plots for each t and corresponding psi_x
  l.set_data(np.linspace(x0,xN,N+1), psi_xt[i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))         #use animation module from matplotlib to create animation
plt.close()                 #remove workspace of matplotlib (duplicate frame)

HTML(ani.to_jshtml())       #draw animation with nav tools

#Save the video in mp4 format (not necessary to generate embedded animation, but nice to have)
#writervideo = matplotlib.animation.FFMpegWriter(fps=60)
#ani.save(filename='SE_CrankNicolson.mp4', writer=writervideo)
```

The initial graph looks like an oscillating gaussian centered at L/2, which makes sense from the formula of $\psi(x,0)$. The effects of the parameters analytically are also what is seen in the graph: \\
*   $\sigma$ changes the width of the gaussian distribution, increasing makes it wider \\
*   $\kappa$ changes the frequency of oscillations given by the complex exponential, increasing it makes it faster \\
*   $x_0$ sets the center of the graph \\

The propagation of this wave is almost uniform to the right, decaying slightly and getting wider, until it collides and becomes a bit chaotic as it reflects off the wall (i.e. its momentum flips, since it is confined to the inside of the box). After the wave settles after the first collision, it is noticeably wider and smaller.

The inverse relationship between wave width and height is explained by the normalization of the wave function. The uncertainty principle also explains why the wave flattens over time: since the position is very localized initially, its uncertainty in x is low, meaning its uncertainty in p is high, causing it to move at different speeds and spreading out. As the uncertainty in x gets higher, the p becomes more certain, causing it to spread out slower.

Notably, the sign of the complex exponential $e^{\pm i\kappa x}$ dictates the direction of the wave's momentum. Setting it to be negative results in the wave initially moving left. This is because applying the momentum operator $\hat{p} = -i\hbar\frac{\partial}{\partial x}$ to $\psi$ will return something proportional to $-\psi$, meaning the momentum eigenvalues are negative.
